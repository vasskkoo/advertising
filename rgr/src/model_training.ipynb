```python
# Імпортуємо необхідні бібліотеки
import pandas as pd
import joblib
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Зчитуємо тестові дані
X_test = pd.read_csv("/Users/mac/Desktop/rgr/data/new_adv.csv")
y_test = X_test.pop("Sales")
```

```python
# Завантажуємо натреновану модель
model = joblib.load("../models/regression_model.pkl")

# Проводимо передбачення на тестових даних
y_pred = model.predict(X_test)

# Оцінюємо результати моделі
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

```python
# Візуалізація результатів
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred)
plt.xlabel("Справжні значення")
plt.ylabel("Прогнозовані значення")
plt.title("Графік справжніх значень проти прогнозованих значень")
sns.regplot(y_test, y_pred, scatter=False, color='red', label='Лінія регресії')
plt.show()
```
