```python
# Імпортуємо необхідні бібліотеки
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Зчитуємо вхідний датасет
data = pd.read_csv("/Users/mac/Desktop/rgr/data/Advertising.csv")
```

```python
# Виводимо перші декілька рядків датасету, щоб побачити його структуру
print("Перші декілька рядків датасету:")
print(data.head())
```

```python
# Розмір датасету (кількість рядків та стовпців)
print("\nРозмір датасету:", data.shape)

# Описова статистика для числових ознак
print("\nОписова статистика для числових ознак:")
print(data.describe())
```

```python
# Перевірка на пропущені значення
print("\nКількість пропущених значень за стовпцями:")
print(data.isnull().sum())
```

```python
# Загальна інформація про датасет
data.info()
```

```python

# Візуалізація розподілу цільової змінної
plt.figure(figsize=(10, 6))
plt.hist(data['Sales'], bins=30, color='skyblue', edgecolor='black')
plt.title("Розподіл цільової змінної")
plt.xlabel("Значення цільової змінної")
plt.ylabel("Частота")
plt.show()
```

```python
# Візуалізація кореляційної матриці для числових ознак
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Кореляційна матриця")
plt.show()
```
